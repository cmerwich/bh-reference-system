```python
def analyse_pgn():
    
    doc_id_list = []
    pgn_verbs_dict = defaultdict(lambda: defaultdict(int))
    pgn_suffix_dict = defaultdict(lambda: defaultdict(int))
    pgn_prps_dict = defaultdict(lambda: defaultdict(int))
    pgn_prde_dict = defaultdict(lambda: defaultdict(int))
    art_dict = collections.Counter() #collections.defaultdict(int)
    
    for book in F.otype.s('book'):
        book_name = T.bookName(book)
        doc_id = book_name
        doc_id_list.append(doc_id)
        
        for ch in L.d(book, 'chapter'):
            chapter = str(F.chapter.v(ch))
            for v in L.d(ch, 'verse'):
                verse = str(F.verse.v(v))
                #doc_id = book_name + chapter + '_' + verse
                #doc_id_list.append(doc_id)

                for w in L.d(v, 'word'):
                    
                    prs = F.prs.v(w) # pronominal suffix -consonantal-transliterated
                    pdp = F.pdp.v(w)   # phrase dependent part-of-speech
                    lex = F.lex.v(w) # lexeme
                    
                    # if the word is a personal pronoun
                    if pdp == 'prps':
                        
                        if lex in prps_dict:
                            pgn_prps_dict[doc_id][prps_dict[lex][0]] += 1
                            for prps in prps_dict.values():
                                if prps[0] not in pgn_prps_dict[doc_id]:
                                    pgn_prps_dict[doc_id][prps[0]] = 0
                    
                    # if the word is a demonstrativum
                    if pdp == 'prde':
        
                        if lex in prde_dict:
                            pgn_prde_dict[doc_id][prde_dict[lex][0]] += 1
                            for prde in prde_dict.values():
                                if prde[0] not in pgn_prde_dict[doc_id]:
                                    pgn_prde_dict[doc_id][prde[0]] = 0

                    # if the word is a verb
                    if pdp == 'verb':
                        
                        pgn_verbs_dict[doc_id][pgn_dict[converse_pgn(F, w)][0]] += 1
                        for pgn in pgn_dict.values():
                            if pgn[0] not in pgn_verbs_dict[doc_id]:
                                pgn_verbs_dict[doc_id][pgn[0]] = 0
                    
                    # if the word is a article
                    if pdp == 'art':
                        
                        art_dict[doc_id] += 1
                    
                    # if the word is a pronominal suffix
                    if prs not in {'n/a', 'absent'}:
                        
                        pgn_suffix_dict[doc_id][suffix_dict[prs][0]] += 1
                        for suffix in suffix_dict.values():
                            if suffix[0] not in pgn_suffix_dict[doc_id]:
                                pgn_suffix_dict[doc_id][suffix[0]] = 0
                    
    #info('PREPARING {} {} {} {} entries in dictionaries'.format(len(pgn_prps_dict), len(pgn_prde_dict), len(pgn_verbs_dict), len(pgn_suffix_dict)))
    #pprint(pgn_verbs_dict)
    return doc_id_list, pgn_prps_dict, pgn_prde_dict, pgn_verbs_dict, art_dict, pgn_suffix_dict
#analyse_pgn()
```

```python
# Step 1: fill in the right dict for PGN_ALL_BOOKS_DICT
# Step 2: fill in the right set for PGN_INFO_LIST
# Step 3: fill in the right filename for FILENAME

doc_id_list, pgn_prps_dict, pgn_prde_dict, pgn_verbs_dict, art_dict, pgn_suffix_dict = analyse_pgn()

PGN_ALL_BOOKS_DICT = pgn_prps_dict # Choose from: pgn_prps_dict, pgn_prde_dict, pgn_verbs_dict, pgn_suffix_dict
PGN_INFO_LIST = sorted(prps_set) # Choose from: prps_set, prde_set, pgn_verbs_set, suffix_set 
FILENAME = 'count_all_pgn_prps_references_hb.csv'

with open(FILENAME, 'w') as f:
    
    header = ['bible_book']
    for info in PGN_INFO_LIST:
        header.append(info)
    f.write('{}\n'.format(','.join(header)))

    for doc in doc_id_list:
        feat_list = [doc]
        for ref in PGN_INFO_LIST:
            if ref in PGN_ALL_BOOKS_DICT[doc]:
                feat_list.append(PGN_ALL_BOOKS_DICT[doc][ref])
            else:
                feat_list.append('0') # bible_book Nahum doesn't contain the feature prde, so add 0
        feat_list = [str(item) for item in feat_list]
        
        f.write('{}\n'.format(','.join(feat_list)))
        
        #pprint(feat_list) # double check generated file with pprint of feat_list
```
